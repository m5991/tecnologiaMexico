```python
# Importación de librerias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
import seaborn as sns

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
from sklearn.tree import export_graphviz
from sklearn.tree import export_text
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
```

```python
# Carga de dataset con informacion 

datosR = pd.read_csv("../data/residente_nivel_max_limpio.csv")
datosR
```

```python
# El nivel de estudios se estandariza de acuerdo las notas del inicio (0: Nada, 1:Kinder, etc). Esto permite una interpración
# más clara de la información y que al momento de crear los predictores la secuencia sea lineal ascendente

datosR['NIVEL']=datosR['NIVEL'].map({0:0,
                                    1:1, 
                                    2:2, 
                                    3:3, 
                                    4:5,
                                    5:3,
                                    6:4,
                                    8:5,
                                    9:6,
                                    10:6,
                                    11:7,
                                    7:99,
                                    99:99}).astype(int)
datosR.head(30)
```

```python
# Carga de dataset

datosV = pd.read_csv("../data/vivienda_hogar_limpio.csv")
datosV
```

```python
# Las tablas las unimos por medio del atributo "code" para tener una sola y la llamamos "datos".

datos = pd.merge(datosR, datosV, on='code')
datos.head()
```

```python
# La encuesta en la que se basa este dataset tiene la peculiaridad de que utilizan el valor 2 como "no", así
# que lo cambiamos por 0 para que la interpretación sea mas logica con 0 = No y 1 = Si


datos['P3_9_1']=datos['P3_9_1'].map({1:1,2:0,9:9}).astype(int)
datos['P3_9_2']=datos['P3_9_2'].map({1:1,2:0,9:9}).astype(int)
datos['P3_9_3']=datos['P3_9_3'].map({1:1,2:0,9:9}).astype(int)
datos['P1_5_1']=datos['P1_5_1'].map({1:1,2:0}).astype(int)
datos['P1_5_2']=datos['P1_5_2'].map({1:1,2:0}).astype(int)
datos['P1_5_3']=datos['P1_5_3'].map({1:1,2:0}).astype(int)
datos['P2_2']=datos['P2_2'].map({1:1,2:0}).astype(int)
datos['P4_1_1']=datos['P4_1_1'].map({1:1,2:0}).astype(int)
datos['P4_1_2']=datos['P4_1_2'].map({1:1,2:0}).astype(int)
datos['P4_1_3']=datos['P4_1_3'].map({1:1,2:0}).astype(int)
datos['P4_1_4']=datos['P4_1_4'].map({1:1,2:0}).astype(int)
datos['P4_1_5']=datos['P4_1_5'].map({1:1,2:0}).astype(int)
datos['P4_1_6']=datos['P4_1_6'].map({1:1,2:0}).astype(int)
datos['P4_2_1']=datos['P4_2_1'].map({1:1,2:0}).astype(int)
datos['P4_2_2']=datos['P4_2_2'].map({1:1,2:0}).astype(int)
datos['P4_4']=datos['P4_4'].map({1:1,2:0}).astype(int)
```

```python
print(datosR.shape)
print(datosV.shape)
print(datos.shape)
```

```python
# Limpieza de nulos, en este caso los valores 9 se consideran nulos ya que son valores donde las personas contestaron que no saben


datos = datos[datos['NIVEL'] != 99]
datos = datos[datos['P3_9_1'] != 9]
datos = datos[datos['P3_9_2'] != 9]
datos = datos[datos['P3_9_3'] != 9]
datos.head(30)
```

```python
datos.describe()
```

```python
# Graficamos el nivel educativo donde podemos ver la distribución de los niveles. La mayor cantidad de personas 
# donde la licenciatura es la que tiene mayor cantidad de registros.
sns.histplot(datos, x="NIVEL")
```

```python
# Hacemos un analisis rapido de los tipos de datos, en particular la variable DOMINIO tiene dos valores
# U para urbano y R para rural.



datos.info()
datos['DOMINIO'].head()
```

```python
# Mapeamos la variable DOMINIO, Urbano valor 1 y Rural valor 0

datos['DOMINIO']=datos['DOMINIO'].map({'U':1,'R':0}).astype(int)
```

```python
datos['DOMINIO'].head()
```

```python
# Con el dataset limpio y organizado podemos hacer el modelo. Dividimos el dataset en sets de entrenamiento
# y de pruebas, donde el objetivo es el Atributo Nivel. El valor random_state es un valor aleatorio que 
# dejamos fijo para poder replicar los experimentos.

X_train, X_test, y_train, y_test = train_test_split(datos.drop(['NIVEL'], axis=1),datos['NIVEL'], random_state=123)

# Creación del modelo
# ------------------------------------------------------------------------------
modelo = DecisionTreeRegressor(
            max_depth         = 5,
            random_state      = 123
          )

# Entrenamiento del modelo
# ------------------------------------------------------------------------------
modelo.fit(X_train, y_train)
```

```python
# Tomamos el modelo para graficar el arbol de desicion. En esta versión observamos los siguientes resultados
# 1. El uso de equipo de computo es el primer determinante (nodo).
# 2. Despues el Grado al que llegaron es el siguiente determinante.
# 3. El Grado se superpone al de Nivel así que lo quitaremos.



fig, ax = plt.subplots(figsize=(12, 5))

print(f"Profundidad del árbol: {modelo.get_depth()}")
print(f"Número de nodos terminales: {modelo.get_n_leaves()}")

plot = plot_tree(
            decision_tree = modelo,
            feature_names = datos.drop(columns = "NIVEL").columns,
            class_names   = 'NIVEL',
            filled        = True,
            impurity      = False,
            fontsize      = 6,
            precision     = 2,
            ax            = ax
       )
```

```python
# El arbol en formato de texto para analizarlo.
texto_modelo = export_text(
                    decision_tree = modelo,
                    feature_names = list(datos.drop(columns = "NIVEL").columns)
               )
print(texto_modelo)
```

```python
# Hacemos un ajuste eliminando variables que no son determinantes
# "code" no tiene un valor importante mas alla de identificar el registro
# "Unnamed: 0_x y 0_y" no se deben utilizar ya que son solamente consecutivos
# Los valores de entidad y UPM no aportan la información suficiente al modelo
# Grado ya habiamos observado que se superpone con el Nivel

# Despues de crear el modelo, obtenemos las siguients conclusiones
# 1. El atributo con mayor peso en el nivel de estudios es el Uso de computadora - estan muy relacionados
# 2. Despues el uso de internet
# 3. en el siguiente nivel es cuantas personas viven en la casa
# 4. En los siguientes niveles estan la edad, el parentesco y el estrato.
# Con esta información podemos concluir que el uso de tecnologias de informacion (computadora, internet, celular),
# esta directamente relacionado con el nivel de estudios. Cuando hacemos una predicción del nivel de estudios,
# el primer parametro a considerar es si los hogares cuentan con servicios de comunicacion (internet sobre todo)


datosX = datos.drop(['code','Unnamed: 0_x','Unnamed: 0_y','GRADO', 'UPM_DIS', 'EST_DIS', 'ENT_y', 'ENT_x'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(datosX.drop(['NIVEL'], axis=1),datosX['NIVEL'], random_state=123)

# Creación del modelo
# ------------------------------------------------------------------------------
modelo = DecisionTreeRegressor(
            max_depth         = 5,
            random_state      = 123
          )

# Entrenamiento del modelo
# ------------------------------------------------------------------------------
modelo.fit(X_train, y_train)

texto_modelo = export_text(
                    decision_tree = modelo,
                    feature_names = list(datosX.drop(columns = "NIVEL").columns)
               )
print(texto_modelo)
```

```python
# Desplegamos que tanta importancia tiene cada uno de los predictores dentro del modelo


importancia_predictores = pd.DataFrame(
                            {'predictor': datosX.drop(columns = "NIVEL").columns,
                             'importancia': modelo.feature_importances_}
                            )
print("Importancia de los predictores en el modelo")
print("-------------------------------------------")
# importancia_predictores.sort_values('importancia', ascending=False)
```

```python
# Hacemos un recorte de ramas para mejorar el desempeño

# Pruning (const complexity pruning) por validación cruzada
# ------------------------------------------------------------------------------
# Valores de ccp_alpha evaluados
param_grid = {'ccp_alpha':np.linspace(0, .25, 10)}

# Búsqueda por validación cruzada
grid = GridSearchCV(
        # El árbol se crece al máximo posible para luego aplicar el pruning
        estimator = DecisionTreeRegressor(
                            max_depth         = None,
                            min_samples_split = 2,
                            min_samples_leaf  = 1,
                            random_state      = 123
                       ),
        param_grid = param_grid,
        cv         = 10,
        refit      = True,
        return_train_score = True
      )

grid.fit(X_train, y_train)

fig, ax = plt.subplots(figsize=(6, 3.84))
scores = pd.DataFrame(grid.cv_results_)
scores.plot(x='param_ccp_alpha', y='mean_train_score', yerr='std_train_score', ax=ax)
scores.plot(x='param_ccp_alpha', y='mean_test_score', yerr='std_test_score', ax=ax)
ax.set_title("Error de validacion cruzada vs hiperparámetro ccp_alpha");
```

```python
# Mejor valor ccp_alpha encontrado
# ------------------------------------------------------------------------------
grid.best_params_
```

```python
# Estructura del árbol final
# ------------------------------------------------------------------------------
modelo_final = grid.best_estimator_

print(f"Profundidad del árbol: {modelo_final.get_depth()}")
print(f"Número de nodos terminales: {modelo_final.get_n_leaves()}")

fig, ax = plt.subplots(figsize=(12, 5))
plot = plot_tree(
            decision_tree = modelo_final,
            feature_names = datosX.drop(columns = "NIVEL").columns,
            class_names   = 'NIVEL',
            filled        = True,
            impurity      = False,
            fontsize      = 10,
            precision     = 2,
            ax            = ax
       )
```

### Predicción y evaluación del modelo

```python
# Error de test del modelo inicial
#-------------------------------------------------------------------------------
predicciones = modelo.predict(X = X_test)

rmse = mean_squared_error(
        y_true  = y_test,
        y_pred  = predicciones,
        squared = False
       )
print(f"El error (rmse) de test es: {rmse}")
```

```python
# Error de test del modelo final (tras aplicar pruning)
#-------------------------------------------------------------------------------
predicciones = modelo_final.predict(X = X_test)

rmse = mean_squared_error(
        y_true  = y_test,
        y_pred  = predicciones,
        squared = False
       )
print(f"El error (rmse) de test es: {rmse}")
```
